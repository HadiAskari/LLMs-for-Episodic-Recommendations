```python
import os
import openai
import pandas as pd
import tiktoken
import json
import requests
import pickle as pkl
from collections import defaultdict
from googlesearch import search
from fake_useragent import UserAgent
from tqdm.auto import tqdm
from bs4 import BeautifulSoup
import re
from inference import main
import ast
```

```python
df_RT_data=pd.read_csv('GAL_500_Crawl.csv', encoding='unicode_escape')
df_GAL_500=pd.read_excel('releaseTestDataset_20230701__GAL500.xlsx')
```

```python
df_RT_data=df_RT_data.fillna('-')
```

```python
def clean(text):  #cleaning RT data collected from extra paragraphs
    text=text.strip().split('\n')[0]
    return text
```

```python
df_RT_data['rt_synopsis']=df_RT_data['rt_synopsis'].apply(clean)
```

```python
def lower(x):
    x=x.lower()
    return x
```

```python
df_RT_data['episode']=df_RT_data['episode'].apply(lower)
df_RT_data['series']=df_RT_data['series'].apply(lower)
```

```python
df_GAL_500['EPISODE TITLE']
```

```python
gal_episodes=df_GAL_500['EPISODE TITLE'].to_list()
gal_series=df_GAL_500['SERIES TITLE'].to_list()

rt_episodes=df_RT_data['episode'].to_list()
rt_series=df_RT_data['series'].to_list()
rt_synopsis=df_RT_data['rt_synopsis'].to_list()
```

```python
count=0
series_titles=[]
episode_titles=[]
synopsis=[]



for ep, ser, syn in zip(rt_episodes,rt_series, rt_synopsis):
    if ep in gal_episodes and ser in gal_series:
        series_titles.append(ser)
        episode_titles.append(ep)
        synopsis.append(syn)
        count+=1

print(count)
```

```python
#inference.py requires a list of series titles, episode titles and a list synopsis (either RT, merlin etc)
res_dic=main(series_titles,episode_titles,synopsis)
```

```python
#save result to file
with open('GAL_Dic_extended.pkl', 'wb') as f:
    pkl.dump(res_dic,f)
```

```python
print(res_dic)
```

```python
df=pd.DataFrame.from_dict(res_dic)
```

```python
df.columns
```

```python
#Code for Recall @ x calculation

for series, episode in zip(series_titles,episode_titles):
    try:
        file_name=df_GAL_500[(df_GAL_500['EPISODE TITLE']==episode) & (df_GAL_500['SERIES TITLE']==series)]['uniqueId'].to_list()[0]
        df_int=df[(df['Episode Title']==episode) & (df['Series Title']==series)]
        res=[]
        for row in df_int.itertuples():
            # print(row)
            dic={"Keyword":row._4, "relevance":row._6, "relevanceClues": row._5}
            res.append(dic)
        json_dic={"result":res}

        with open("output_json/{}.json".format(file_name), "w") as outfile:
            json.dump(json_dic, outfile)



    except Exception as e:
        print(e)
        pass
```

```python
#Code for Recall @ x calculation

total_count=0
suc_count=0
for series, episode in zip(series_titles,episode_titles):
    try:
        df_int=df[(df['Episode Title']==episode) & (df['Series Title']==series)]
        actual_raw=df_GAL_500[(df_GAL_500['EPISODE TITLE']==episode) & (df_GAL_500['SERIES TITLE']==series)]['adWordList'].values
        actual=ast.literal_eval(actual_raw[0])
        actual=[x.strip() for x in actual]
        print(actual)
        total_count+=len(actual)
        intermediate=0
        for row in df_int.itertuples():
            if float(row._6)>=0.1 and row._4 in actual:
                suc_count+=1
                intermediate+=1
        
        if intermediate==0:
            print(df_GAL_500[(df_GAL_500['EPISODE TITLE']==episode) & (df_GAL_500['SERIES TITLE']==series)]['uniqueId'])
        
        
    except Exception as e:
        print(e)
        pass

print(suc_count/total_count)
```
